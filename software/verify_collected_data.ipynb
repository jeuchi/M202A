# Visualizing data collected using the CARLA Simulator

```python
%matplotlib inline

import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
import math
from PIL import Image
```

Load .npy data 

```python
data = np.load('../out/CameraSeg/0.npy')

print(f'Shapes: Semantic segmentation ground truth: {data.shape}')
print(f'Data types: Semantic segmentation ground truth: {data.dtype}')
```

Plotting images

```python
def plot_images(arr, offset, num_images):
    # Create a figure and axis for plotting
    plt.figure(figsize=(15, 10))  # Adjust the figure size as needed
    divisions = int(math.sqrt(num_images))

    # Loop through the data and plot each image
    for i in range(num_images):
        plt.subplot(divisions, divisions, i + 1)  # Create a grid of 10x10 subplots, adjust as needed
        plt.imshow(arr[i+offset], cmap='tab20', aspect='auto')  # Use 'cmap' according to your data
        plt.axis('off')  # Turn off axis labels

    plt.tight_layout()  # Ensure that the subplots don't overlap
    plt.show()
```

```python
plot_images(data, 750, 1)
```
